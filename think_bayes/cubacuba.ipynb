```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
```

```python
pop: np.ndarray = np.random.normal(0, 1, 10_000)
sam30: np.ndarray = np.random.choice(pop, 30)
```

```python
sam6x5: np.ndarray = np.empty(shape=5)

for i in range(5):
    sam6x5[i] = np.mean(np.random.choice(sam30, 6, replace=False))
```

```python
sam20: np.ndarray = np.random.choice(pop, 20)

sam4x5: np.ndarray = np.empty(shape=5)

for i in range(5):
    sam4x5[i] = np.mean(np.random.choice(sam20, 4, replace=False))


sns.kdeplot(sam20)
sns.kdeplot(sam4x5)
```

```python
print(pop.mean())
print(pop.std())
print(f'{pop.mean() - 2. * pop.std()} - {pop.mean() + 2. * pop.std()}')
print(stats.norm.interval(0.95, loc=np.mean(pop), scale=stats.sem(pop)))
```

```python
print(sam30.mean())
print(sam30.std())
print(f'{sam30.mean() - 2. * sam30.std()} - {sam30.mean() + 2. * sam30.std()}')
print(stats.norm.interval(0.95, loc=np.mean(sam30), scale=stats.sem(sam30)))
```

```python
print(sam6x5.mean())
print(sam6x5.std())
print(f'{sam6x5.mean() - 2. * sam6x5.std()} - {sam6x5.mean() + 2. * sam6x5.std()}')
print(stats.norm.interval(0.95, loc=np.mean(sam6x5), scale=stats.sem(sam6x5)))
```

```python
sns.kdeplot(pop)
sns.kdeplot(sam30)
sns.kdeplot(sam6x5)
```

```python
hbsag: np.ndarray = np.random.normal(0.6, 0.2, 1_000)
sns.histplot(hbsag, kde=True)
```

```python
p = hbsag.mean()
s = hbsag.std()
a = p - 3 * s
b = p - 2 * s
c = p - s
d = p
e = p + s
f = p + 2 * s
g = p + 3 * s
graf_v = sns.kdeplot(hbsag)
graf_v.vlines([a, b, c, d, e, f, g], 0, graf_v.get_ylim()[1], colors="0.75")
```

```python
graf_h = sns.kdeplot(y=hbsag)
graf_h.hlines([a, b, c, d, e, f, g], 0, graf_h.get_xlim()[1], colors="0.75")
```

```python
pop_cerun = np.random.normal(0., 0.2, 1_000)
sam_cerun = np.random.choice(pop_cerun, 10)

b, a = stats.norm.interval(0.95, loc=np.mean(sam_cerun), scale=stats.sem(sam_cerun))
graf_cerun = sns.kdeplot(sam_cerun)
graf_cerun.vlines([a, b, 0.], 0, graf_cerun.get_ylim()[1], colors="0.75")
```

```python
pop1_ltlv = np.random.normal(5.0, 1.0, 1_000)
pop2_ltlv = np.random.normal(7.0, 1.0, 1_000)

sam3_lot1 = np.random.choice(pop1_ltlv, 3)
sam3_lot2 = np.random.choice(pop2_ltlv, 3)

p, q = stats.norm.interval(0.95, loc=np.mean(sam3_lot1), scale=stats.sem(sam3_lot1))
r, s = stats.norm.interval(0.95, loc=np.mean(sam3_lot2), scale=stats.sem(sam3_lot2))

graf_l1 = sns.kdeplot(sam3_lot1, color="b")
graf_l2 = sns.kdeplot(sam3_lot2, color="r")
graf_l1.axvspan(p, q, color="b", alpha=0.1)
graf_l2.axvspan(r, s, color="r", alpha=0.1)
```

```python
pop1_ltlv = np.random.normal(5.0, 1.0, 1_000)
pop2_ltlv = np.random.normal(6.0, 1.0, 1_000)

sam3_lot1 = np.random.choice(pop1_ltlv, 10)
sam3_lot2 = np.random.choice(pop2_ltlv, 10)

p, q = stats.norm.interval(0.95, loc=np.mean(sam3_lot1), scale=stats.sem(sam3_lot1))
r, s = stats.norm.interval(0.95, loc=np.mean(sam3_lot2), scale=stats.sem(sam3_lot2))

graf_l1 = sns.kdeplot(sam3_lot1, color="b")
graf_l2 = sns.kdeplot(sam3_lot2, color="r")
graf_l1.axvspan(p, q, color="b", alpha=0.1)
graf_l2.axvspan(r, s, color="r", alpha=0.1)
```
