```python
import pandas as pd

from fractions import Fraction as frac
```

Mangkuk 1: 30 vanila, 10 coklat  
Mangkuk 2: 20 vanila, 20 coklat

P(m1|v)?

```python
# P(v|m1) * P(m1) / P(v)
frac(30, (30+10)) * frac(1, 2) / frac((30+20), (30+10+20+20))
```

Pengiraan dengan menggunakan jadual Bayes

```python
data: dict = {
    "mangkuk": ["m1", "m2"],
    "prior": [frac(1, 2), frac(1, 2)],
    "likelihood": [frac(30, 40), frac(20, 40)]
}

jadual_bayes: pd.DataFrame = pd.DataFrame(data)
jadual_bayes
```

```python
def kemaskini(jad) -> float:
  jad["pengangka"] = jad["prior"] * jad["likelihood"]
  prob_data: float = jad["pengangka"].sum()
  jad["posterior"] = jad["pengangka"] / prob_data
  return prob_data
```

```python
penyebut: float = kemaskini(jadual_bayes)

jadual_bayes
```

dadu: d6, d8, d12  
P(nombor=1)?

```python
dadu: dict = {
    "dadu": ["d6", "d8", "d12"],
    "prior": [frac(1, 3), frac(1, 3), frac(1, 3)],
    "likelihood": [frac(1, 6), frac(1, 8), frac(1, 12)]
}


jadual_bayes_2: pd.DataFrame = pd.DataFrame(dadu)

penyebut_dadu: float = kemaskini(jadual_bayes_2)

print(jadual_bayes_2)
print("*"*80)
print(f'{penyebut_dadu = }')
```

The Monty Hall Problem  
Mula-mula pilih pintu a, Monty buka pintu c

```python
monty: dict = {
    "door": ["a", "b", "c"],
    "prior": [frac(1, 3), frac(1, 3), frac(1, 3)],
    "likelihood": [frac(1, 2), 1, 0]
}

jadual_bayes_3: pd.DataFrame = pd.DataFrame(monty)

kemaskini(jadual_bayes_3)

jadual_bayes_3
```
