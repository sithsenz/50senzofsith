```python
# @title # Bland-Altman Plot {"vertical-output":true,"display-mode":"form"}


# @markdown Sila masukkan nilai-nilai x dan y dalam ruangan berikut dengan dipisahkan oleh simbol koma ",".


# @markdown Contoh: 1.0 , 2.0 , 3.0
x_input = "53, 54, 52, 76, 50, 63, 80, 82"  # @param {type: "string"}
y_input = "59, 56, 56, 83, 56, 70, 87, 85"  # @param {type: "string"}


# @markdown Sila klik pada menu `Runtime` &rarr; `Run all` selepas selesai memasukkan nilai-nilai x dan y.


X = [eval(x) for x in x_input.split(',')]
Y = [eval(y) for y in y_input.split(',')]


if len(X) == len(Y):
    blandAltman(Y, X, confidenceInterval=None)
else:
    mesej: str = f'Bilangan data x_input dan y_input harus sama. x_input mengandungi {len(X)} data, manakala y_input mengandungi {len(Y)} data.'
    raise ValueError(mesej)
```
